## Preparation of Data
Required library imported.

```r
library(xts)
```

Data in csv file is stored in cons.

```r
cons <- read.csv("RealTimeConsumption-01012016-26042020.csv")
```

```r
head(cons)
summary(cons)
```

```r
cons$Date <- strptime(as.character(cons$Date), "%d.%m.%Y")
```

```r
cons$datetime = with(cons, as.POSIXct(paste(Date, Hour) ) )
cons = cons[c("datetime","Consumption..MWh.")]
names(cons) = c("datetime","consumption")
```

```r
cons$consumption <- gsub(",","",cons$consumption)
cons$consumption <- as.numeric(cons$consumption)
```

```r
str(cons)
head(cons)
summary(cons)
```

```r
cons <- cons[order(cons$datetime),]
```

```r
head(cons)
```

```r
consxts = xts(cons$consumption,order.by = cons$datetime)
```

```r
consts<-ts(consxts,start=2016,freq=365*24)
```

```r
ts.plot(consts,xlab = "Time", ylab = "Consumption in MWh",
        main="Figure 1 - Hourly Electricity Consumption in Turkey")
```

```r
str(consts)
```

Mean of consumptions of days, weeks, months and quarters are calculated. Mean preferred against sum because of the possible different lengths of periods.

```r
consts_day=ts(apply.daily(consxts,FUN=mean),start=2016,freq=365)
consts_week=ts(apply.weekly(consxts,FUN=mean),start=2016,freq=52)
consts_month=ts(apply.monthly(consxts,FUN=mean),start=2016,freq=12)
consts_quarter=ts(apply.quarterly(consxts,FUN=mean),start=2016,freq=4)
```

```r
ts.plot(consts_day,xlab = "Days", ylab = "Avg. Hourly Consumption per Day (MWh)",
        main="Figure 2 - Avg. Hourly Electricity Consumption per Day")
```

```r
ts.plot(consts_week,xlab = "Weeks", ylab = "Avg. Hourly Consumption per Week (MWh)",
        main="Figure 3 - Avg. Hourly Electricity Consumption per Week")
```

```r
ts.plot(consts_month,xlab = "Months", ylab = "Avg. Hourly Consumption per Month (MWh)",
        main="Figure 4 - Avg. Hourly Electricity Consumption per Month")
```

```r
ts.plot(consts_quarter,xlab = "Quarters", ylab = "Avg. Hourly Consumption per Quarter (MWh)",
        main="Figure 5 - Avg. Hourly Electricity Consumption per Quarter")
```

It is predicted that when consumption increase, seasonality will become wider. That's why, type is selected as multiplicative.

```r
dec_quarter<-decompose(consts_quarter,type="multiplicative")
plot(dec_quarter)
```

```r
dec_month<-decompose(consts_month,type="multiplicative")
plot(dec_month)
```

```r
dec_week<-decompose(consts_week,type="multiplicative")
plot(dec_week)
```

```r
dec_day<-decompose(consts_day,type="multiplicative")
plot(dec_day)
```

```r
dec<-decompose(consts,type="multiplicative")
plot(dec)
```

```r
acf(consts,lag.max = 400, main="ACF of Hourly Consumption")
```

Strong autocorrelation is observed in 1 day (24 hours) and 7 days (168 hours) periods.

```r
remove_season<-consts/dec$seasonal
ts.plot(remove_season, main="Figure 6 - Comsumption without Seasonality")
acf(remove_season,lag.max = 400, main="ACF when Seasonality Removed", na.action = na.pass)
```

```r
remove_season_trend<-remove_season/dec$trend
ts.plot(remove_season_trend, main="Figure 7 - Comsumption without Seasonality and Trend")
acf(remove_season_trend,lag.max = 400, main="ACF when Seasonality and Trend Removed", na.action = na.pass)

```

```r
aics = c(1:20)*0
for (i in 0:14){
    ar<-arima(remove_season_trend,order=c(i,0,0))
    aics[i+1]=ar$aic
}
```

After 12, AIC does not decrease significantly, 12 (aic = -177431.38) has chosen.


```r
aics2 = c(1:10)*0
for (i in 0:9){
    ma<-arima(remove_season_trend,order=c(0,0,i))
    aics2[i+1]=ma$aic
}
```

q=10 (aic = -173275.73) has chosen.

```r
ar<-arima(remove_season_trend,order=c(12,0,0))
```

```r
ma<-arima(remove_season_trend,order=c(0,0,10))
```

```r
ar$aic
```

```r
ma$aic
```

AR model is better because it has lower AIC value.

Model is applied to original data to forecast.

```r
ar<-arima(consts,order=c(12,0,0))
```

```r
ar$aic
```

```r
ar_fitted<- consts-residuals(ar)
```

```r
ar_forecast <- predict(ar, n.ahead = 24)
```

```r
ar_forecast
```

```r
ts.plot(consts,xlab = "Hours", ylab = "Consumption",xlim=c(2020.25,2020.33),main="Figure 8 - Hourly Consumption and 24 Hours Forecast")
points(ar_fitted, type = "l", col = 2)
points(ar_forecast$pred, type = "p", col = 4, lty = 2)
points(ar_forecast$pred - 1.96*ar_forecast$se, type = "l", col = 4, lty = 2)
points(ar_forecast$pred + 1.96*ar_forecast$se, type = "l", col = 4, lty = 2)
```

To observe easily, last part of the data is plotted.
